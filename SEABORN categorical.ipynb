```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='tip',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips,color='black')
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.factorplot(x='tip',data=tips,kind='count')
```

```python
sns.factorplot(x='day',y='tip',data=tips,kind='box')
```
